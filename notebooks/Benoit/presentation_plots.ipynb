```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter
```

## Introduction plot

Source : 
- https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5824202/
- https://www.ncoa.org/adviser/hearing-aids/hearing-loss-america/ 

```python
# Data
# https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5824202/
years = ['2020', '2030', '2040', '2050', '2060']
age_groups = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '≥80']

# Moderate or greater hearing loss data
moderate_data = {
    '20-29': [0.03, 0.03, 0.03, 0.03, 0.03],
    '30-39': [0.30, 0.32, 0.32, 0.33, 0.35],
    '40-49': [0.19, 0.21, 0.23, 0.23, 0.24],
    '50-59': [1.18, 1.12, 1.27, 1.38, 1.40],
    '60-69': [2.61, 2.67, 2.56, 2.95, 3.22],
    '70-79': [4.75, 6.43, 6.66, 6.52, 7.61],
    '≥80': [6.02, 9.03, 12.98, 14.82, 15.45]
}

# Calculate cumulative values
moderate_cumulative = np.zeros(len(years))

for age_group in moderate_data:
    moderate_cumulative += np.array(moderate_data[age_group])

```

```python
# Plotting
plt.figure(figsize=(10, 6))

plt.plot(years, moderate_cumulative, marker='o', color='#adffb7', linewidth=4, markersize=10)

# Add labels, title
plt.xlabel('Year', color='#f1f1f1', fontsize=18)
plt.ylabel('Population (Millions)', color='#f1f1f1', fontsize=18)
plt.title('Evolution of U.S. Population with Hearing Loss', color='#f1f1f1', fontsize=22)
plt.xticks(years, color='#f1f1f1', fontsize=18)
plt.yticks(color='#f1f1f1', fontsize=18)
plt.tick_params(axis='x', colors='#f1f1f1') # Set tick marker color to white
plt.tick_params(axis='y', colors='#f1f1f1') # Set tick marker color to white

# Set background color
plt.gca().set_facecolor('#f1f1f1')

# Set color of all axes to white
plt.gca().spines['bottom'].set_color('#f1f1f1')
plt.gca().spines['left'].set_color('#f1f1f1')
plt.gca().spines['right'].set_color('#f1f1f1')
plt.gca().spines['top'].set_color('#f1f1f1')

# Format y-axis tick labels to include "M" for million
formatter = FuncFormatter(lambda x, _: f'{x:.0f} M')
plt.gca().yaxis.set_major_formatter(formatter)

# Set y-axis limit close to the current value
plt.ylim(bottom=14)

plt.tight_layout()

# Save as transparent PNG
plt.savefig('plots/moderate_hearing_loss_line_chart.png', transparent=True)

plt.show()
```
